### Max Min Function
Creating a function that gets the max and min of a given varibles and its associated varibles. 

```python
# min and max values functions

def min_max_values(col):
    ''' this function takes the column name as the argument
        and returns the top and bottom obs in the dataframe
    '''
    top = Trading[col].idxmax() #top max index
    top_obs = pd.DataFrame(Trading.loc[top])
    
    bottom = Trading[col].idxmin() #bottom min index
    bot_obs = pd.DataFrame(Trading.loc[bottom])
    min_max_obs = pd.concat([top_obs, bot_obs], axis=1) # then we concatenate the top and bottom together 
    return min_max_obs
```

## Using pairplot from Seabourn package 
We plot the correlation for all of the
numerical varibles with respect to price.
The following loop will loop through all the variable for x axis and while leaving the y axis fixed to price. 

```python
#For loop
for i in range(0, len(Trading_num.columns),5):
    sns.pairplot(Trading_num, y_vars=['Price'],x_vars=Trading_num.columns[i:i+5])
```

## Ploting using a heatmap 
The following plot will show any varibles with a correlation of greater than .05 or less than -0.4.

```python
#plotting significant correlation in one plot using a heatmap
# so any varible showing significant correlation will 
# appear in this heatmap
fig = plt.figure(figsize=(20, 20)) # changes the figure size
corr = Trading_num.drop('Price', axis=1).corr()
sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)],
            cmap='viridis', vmax=1.0, vmin=1.0, linewidth=0.1, 
            annot=True, annot_kws={"size": 8}, square=True);
```
